```python
#Importing required libraries

import seaborn as sns 
import pandas as pd
import numpy as np
```

```python
#Loading dataset from seaborn

df=sns.load_dataset("iris")
```

```python
df.head()
```

```python
#Getting unique values from species column

df['species'].unique()
```

```python
#Finding any null values in the dataset

df.isna()
```

```python
#Removing setosa from dataset 

df=df[df['species']!='setosa']
```

```python
df.head()
```

```python
#Tranforming Categorical data to numerical data 

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['species'] = le.fit_transform(df['species'])
```

```python
df.head()
```

```python
# Split dataset into independent and dependent features 

x=df.iloc[:,:-1]   #Taking all the rows, and all the columns leaving the last column
y=df.iloc[:,-1]    #Taking all the rows, but only the ast column
```

```python
x
```

```python
y
```

```python
#Train test split 

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
#Model initialization 

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#Fitting my dataset into the model

model.fit(x_train, y_train)
```

```python
#Prediction 

y_pred=model.predict(x_test)
y_pred
```

```python
#Accuarcy

from sklearn.metrics import accuracy_score,classification_report

score=accuracy_score(y_pred,y_test)                             # Accuracy = No. of correct Prediction / Total no. of predictions
score
```

```python
print(classification_report(y_pred,y_test))
```
