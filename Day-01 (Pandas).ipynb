**1. Pandas Basics**

```python
# importing pandas library
import pandas as pd
```

```python
# read.csv()
df = pd.read_csv("day1.csv")
df.head()
```

```python
# head and tail

print(df.head())
print(df.tail())
```

```python
# rename()

df_renamed = df.rename(columns={"Name": "FullName",
                                "Score": "ExamScore", 
                                "Department": "Dept"})
df_renamed.head()
```

```python
# info()

df.info()
```

```python
# describe()

df.describe()
```

```python
# loc (label-based selection)

df.loc[0, 'Score']
```

```python
df.loc[0:2, ['Name','Department']]
```

```python
# iloc (index-based selection)

df.iloc[0, 4]
```

```python
df.iloc[0:2, [1, 6]]
```

**2. Data Preprocessing & Cleaning**

```python
# isna / isnull
df.isna()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
# dropna
df_dropped = df.dropna()
df_dropped

# dropna(axis=1) drops column with null value instead of ROW
```

```python
# fillna
df_filled = df.fillna(df['Score'].mean())
df_filled
```
