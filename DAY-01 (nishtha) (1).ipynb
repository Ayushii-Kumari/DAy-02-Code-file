```python
import zipfile 
import os
```

```python
def extract_zip(root, dest):
    os.makedirs(dest, exist_ok = True)    #line will be skipped if the directory is already present 
    with zipfile.ZipFile(root, 'r') as zip_ref:     # MAKE SURE TO USE "F" in ZipFile.   #reads the zip file    # r means read binary , w means write binary 
        zip_ref.extractall(dest)   # extracts the file to the destination 
        
    print("All the files are extracted succesfully")
```

```python
file_path = "datasets_day2.zip"
output_path = os.path.dirname(file_path) if os.path.dirname(file_path)  else"."

extract_zip(file_path, output_path)
```
